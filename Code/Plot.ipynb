```python
import csv
import matplotlib.pyplot as plt
```

```python
res = [1,2,3,4,5,6,7,8]
hit_rate = []
no_hit = []

with open('F:/Service Cache Implementation/Experiments/LRU-SingleResource/RAM/1.1/Results.csv') as csvfile:
    x = csv.reader(csvfile, delimiter=',')

    for row in x:
        no_hit.append(row[2])
        y = round(float(row[3]),2)
        hit_rate.append(y)
```

```python
plt.plot(res,hit_rate, color='g', linestyle='dashed',marker='o')
plt.xlabel('Resources')
plt.ylabel('Cache Hit Rates(%)')
plt.title('Hit Rate vs Resources', fontsize = 16)
plt.grid(axis='y', linestyle='-')
plt.savefig('F:/Service Cache Implementation/Experiments/LRU-SingleResource/RAM/1.1/HitRate vs Resources.png')
plt.show()
```

```python
plt.plot(res,no_hit, color='g', linestyle='dashed',marker='o')
plt.xlabel('Resources')
plt.ylabel('Number of Hits')
plt.title('Number of Hits vs Resources', fontsize = 16)
plt.grid(axis='y', linestyle='-')
plt.savefig('F:/Service Cache Implementation/Experiments/LRU-SingleResource/RAM/1.1/NoHit vs Resources.png')
plt.show()
```
